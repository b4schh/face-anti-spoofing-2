```python
import os
import cv2
from tqdm import tqdm

# ============================
# CẤU HÌNH
# ============================

DOMAIN = "SiW"
ROOT_DIR = "dataset_processed"
IMG_SIZE = 256

LIVE_DIR = f"{ROOT_DIR}/{DOMAIN}/live"
SPOOF_DIR = f"{ROOT_DIR}/{DOMAIN}/not_live"

TARGET_DIRS = [LIVE_DIR, SPOOF_DIR]

print("Folders to resize:")
print(LIVE_DIR)
print(SPOOF_DIR)

# ============================
# HÀM RESIZE ẢNH
# ============================

def resize_images_in_folder(folder, size):
    count = 0
    for fname in tqdm(os.listdir(folder), desc=f"Resizing {os.path.basename(folder)}"):
        path = os.path.join(folder, fname)

        if not fname.lower().endswith((".jpg", ".jpeg", ".png")):
            continue

        img = cv2.imread(path)
        if img is None:
            print("⚠ Không đọc được ảnh:", path)
            continue

        resized = cv2.resize(img, (size, size))
        cv2.imwrite(path, resized)

        count += 1
    return count

# ============================
# CHẠY RESIZE
# ============================

total = 0
for d in TARGET_DIRS:
    if os.path.exists(d):
        total += resize_images_in_folder(d, IMG_SIZE)
    else:
        print("⚠ Folder không tồn tại:", d)

print("\n==============================")
print(f"✓ Resize hoàn tất ({total} ảnh)")
print("==============================")
```
